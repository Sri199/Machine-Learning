### Feature Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
cols = ['City','City Group','Type']
for i in cols:
    df[i] = le.fit_transform(df[i])
```

```python
x = df.drop(['Open Date','revenue','City'], axis = 1)
y = df['revenue']
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

### Train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

### Implementing Models

### LinearRegression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(fit_intercept=True,positive=True,copy_X= True)
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
mean_squared_error(y_true= y_test,y_pred=y_pred)
```

```python
r2_score(y_true = y_test,y_pred = y_pred)
```

### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(x_train, y_train)
```

```python
# mean squred error
rf_pred = rf_regressor.predict(x_test)
mean_squared_error(y_test, rf_pred)
```

```python
# r2_score
r2_score(y_true = y_test,y_pred=rf_pred)
```

### GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor(learning_rate= 0.01, max_depth =  5, n_estimators = 100, subsample = 0.8)
gbr.fit(x_train,y_train)
```

```python
# mse
gb_pred = gbr.predict(x_test)
mean_squared_error(y_true=y_test,y_pred=gb_pred)
```

```python
r2_score(y_true = y_test,y_pred = gb_pred)
```

### DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(max_depth =5, min_samples_leaf = 4,min_samples_split = 10)
dt.fit(x_train,y_train)
```

```python
dt_pred = dt.predict(x_test)
mean_squared_error(y_test,dt_pred)
```

```python
r2_score(y_true = y_test,y_pred=dt_pred)
```
