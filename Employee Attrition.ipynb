### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms
```

```python
data['average_montly_hours'] = mms.fit_transform(data[['average_montly_hours']])
data
```

### Feature Encoding

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False)
```

```python
ohetransform = ohe.fit_transform(data[['department','salary']])
columnames = ohe.get_feature_names_out(['department', 'salary']) 
```

```python
encoded_df = pd.DataFrame(ohetransform, columns=columnames)
encoded_df
```

```python
data = pd.concat([data, encoded_df],axis=1)
```

```python
data
```

### Splitting Train Test Data

```python
X = data.drop(['left'],axis=1)
y = data['left']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=0)
```

```python
X_train
```

```python
y_test
```

### Implementing models

### LinearRegression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(fit_intercept=True,positive=True,copy_X= True)
lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
y_pred = lr.predict(X_test)
mean_squared_error(y_true= y_test,y_pred=y_pred)
```

```python
r2_score(y_true = y_test,y_pred = y_pred)
```

### GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor(learning_rate= 0.01, max_depth =  5, n_estimators = 100, subsample = 0.8)
gbr.fit(X_train,y_train)
```

```python
gb_pred = gbr.predict(X_test)
mean_squared_error(y_true=y_test,y_pred=gb_pred)
```

```python
r2_score(y_true = y_test,y_pred = gb_pred)
```

### DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(max_depth =5, min_samples_leaf = 4,min_samples_split = 10)
dt.fit(X_train,y_train)
```

```python
dt_pred = dt.predict(X_test)
mean_squared_error(y_test,dt_pred)
```

```python
r2_score(y_true = y_test,y_pred=dt_pred)
```

### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, y_train)
```

```python
rf_pred = rf_regressor.predict(X_test)
mean_squared_error(y_test, rf_pred)
```

```python
r2_score(y_true = y_test,y_pred=rf_pred)
```

### From the above 4 models RandomForestRegressor has best accuracy score of 0.93
